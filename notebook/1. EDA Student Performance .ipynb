Define Numerical and catagorical columns

```python
numeric_features = [features for features in df.columns if df[features].dtype != 'str']
catagorical_features = [features for features in df.columns if df[features].dtype == 'str']

print('we have {} numeric features: {}'.format(len(numeric_features),numeric_features))
print('we have {} catagorical features: {}'.format(len(catagorical_features),catagorical_features))
```

## 4) Exploring Data (Visualization)

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(4, 2, figsize=(15, 20))

sns.countplot(data=df, x='Degree Program', ax=axs[0,0])
sns.countplot(data=df, x='CGPA  Average', ax=axs[0,1])

sns.countplot(data=df, x='Internship Experience', ax=axs[1,0])
sns.countplot(data=df, x='Number of Projects Completed', ax=axs[1,1])

sns.countplot(data=df, x='Certifications Completed', ax=axs[2,0])
sns.countplot(data=df, x='DSA / Syllabus Coverage', ax=axs[2,1])

sns.countplot(data=df, x='Aptitude/ Syllabus Coverage', ax=axs[3,0])

# Hide the unused subplot
axs[3,1].axis('off')

for ax in axs.flat:
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
